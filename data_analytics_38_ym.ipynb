```python
import requests
from pprint import pprint
from yaml import load, FullLoader
```

```python
config_file = open('config.yaml', 'r')
config = load(config_file, Loader=FullLoader)
```

```python
token = config['token']
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
counter = '51583145'
startDate = '2020-12-01'
endDate = '2020-12-31'
dimensions = ['ym:s:date', 'ym:s:deviceCategory', 'ym:s:browser']
metrics = ['ym:s:visits', 'ym:s:users', 'ym:s:pageviews', 'ym:s:percentNewVisitors']
```

```python
headers = {'Authorization': 'OAuth ' + token}
url = 'https://api-metrika.yandex.ru/management/v1/counter/{}/grants'.format(counter)
users_grants = requests.get(url, headers=headers).json()
```

```python
users_grants
```

```python
filters = "ym:s:regionCountryName=='Россия'"
```

```python
metrics_string = ','.join(metrics)
print(metrics_string)
```

```python
dimensions_string = ','.join(dimensions)
```

```python
params = {
    'date1': startDate,
    'date2': endDate,
    'id': counter,
    'dimensions': dimensions_string,
    'metrics': metrics_string,
    'filters': filters,
}
```

```python
r = requests.get(API_URL, headers=headers, params = params)
print(r)
data = r.json()
```

```python
for line in data['data']:
    visit_data = line['dimensions'][0]['name']
    visits = line['metrics'][0]
    users = line['metrics'][1]
    pageviews = line['metrics'][2]
    percent_new_visitors = line['metrics'][3]
#     print(f'visit_date: {visit_date}, visits: {visits}, users: {users}, pageviews: {pageviews}, \
#           percent_new_visitors: {percent_new_visitors}')
    print(f'(new_to_all: {round(visits * (percent_new_visitors / 100) / users, 2)}')
```

```python
def make_request(**kwargs):
    """ Запрос к API Яндекс метрики с произворльным
    числом параметров"""
    
    return requests.get(API_URL, headers=headers, params=kwargs).json()
```

```python
filters = "ym:s:regionCountryName=='Россия'"
```

```python
data = make_request(date1=startDate, date2=endDate, dimensions=','.join(dimensions), \
                   metrics=','.join(metrics), filters=filters, id=counter, accuracy=1)
```
