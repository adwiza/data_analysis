```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://lenta.ru/articles/2018/07/01/pobeda/'
```

```python
request_data = requests.get(url)
```

```python
request_data.text
```

```python
soup = BeautifulSoup(request_data.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.title.text
```

```python
soup.find(itemprop="headline").text == soup.find(class_='b-topic__title').text
```

```python
topic = soup.find('div', class_="b-topic__info")
topic.text
```

```python
data = soup.find('time', class_='g-date')
```

```python
data.text
```

```python
news_text = soup.find(class_='b-topic__announce')
```

```python
news_text.text
```

```python
for link in news_text.find_all('a'):
    print(link)
    print(link.get('href')) 
    
```
