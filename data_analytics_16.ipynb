```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
data_pivot = data.pivot_table(index=['movieId', 'rating'], values='timestamp', aggfunc='count').reset_index()
data_pivot.head(10)
```

```python
data_pivot.groupby('movieId').timestamp.transform(lambda x: x)
```

```python
data_pivot['sum'] = data_pivot.groupby('movieId').timestamp.transform(lambda x: sum(x))
data_pivot.head(10)
```

```python
data_pivot['share'] = data_pivot['timestamp'] / data_pivot['sum']
data_pivot.head(10)
```

```python
print(data_pivot[data_pivot['movieId'] == 1]['share'].sum())
```

```python
def rating_classifier(rating):
    if rating <= 2:
        return 'Низкий'
    elif rating <= 4:
        return 'Средний'
    else:
        return 'Высокий'
```

```python
# data_pivot['class'] = data_pivot['rating'].apply(lambda x: 'Низкий' if x <= 2 else ('Высокий' if x > 4 else 'Средний'))
# data_pivot['class'] = data_pivot['rating'].apply(lambda x: rating_classifier(x))
data_pivot['class'] = data_pivot['rating'].apply(rating_classifier)
data_pivot.head(10)
```

```python
print(data_pivot[(data_pivot['movieId'] == 356) & (data_pivot['class'] == 'Средний')]['share'].sum().round(2))
# print(data_pivot[(data_pivot['movieId'] == 356) & (data_pivot['class'] == 'Средний')]['share'].sum())
```
